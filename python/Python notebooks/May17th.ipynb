### Problem 2:
##### Problem statement
Define a function to add a new data to a CSV file.
#### Constraints

#### Test Cases
* Test Case 1
* Test Case 2


```python
def addrowCSV(filename,rowdata):
    df = readCSV(filename) # get dataframe from csv
    df.loc[len(df)] = rowdata # add a row to dataframe
    df.to_csv(filename,index = False) #send dataframe back to csv.
    
    return df
```

```python
rowdata = ['010101','statename',123,4335,4565,8907,5647,2314,4567,7689,9034]
addrowCSV('Data Files/income.csv',rowdata)
```

```python
def updaterowCSV(filename,rowindex,rowdata):
    df = readCSV(filename)
    df.loc[rowindex] = rowdata
    df.to_csv(filename,index = False)
    return df


#df = df.drop(['Unnamed: 0','Unnamed: 0.1'],'axis = 1')
#df
```

### Problem 4:
##### Problem statement
Define a function to delete a row from a CSV file.
#### Constraints

#### Test Cases
* Test Case 1
* Test Case 2


```python
def deleterowCSV(filename,rowindex):
    df = readCSV(filename)
    df = df.drop(rowindex)
    df.to_csv(filename,index = False)
    return df
```

```python
filename = 'Data Files/income.csv'
rowindex = 5
deleterowCSV(filename,rowindex)
```
