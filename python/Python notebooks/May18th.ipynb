### Problem 3:
##### Problem statement
plot a line graph on **income vs year** for the state of california .
#### Constraints

#### Test Cases
* Test Case 1
* Test Case 2


```python
import pandas as pd
df = pd.read_csv('Data Files/income.csv')
y = df.values[4]
x = df.columns
x
```

```python
years = []
for i in range (2,len(x)):
    years.append(x[i])
    
    
years

income = []
for i in range (2,len(y)):
    income.append(y[i])
income


plt.xlabel('Year')
plt.grid(True)
plt.ylabel('Income')
plt.plot(years,income)
plt.title('Income of california state')
plt.savefig('python figures/income.png')

plt.show
```

# Bar plot

```python
width = 1/2
plt.title('Income of California state')
plt.bar(years,income,width,color = 'grey' )   #png formate is better as it gives more clarity.(portable network graphics)
plt.savefig('python figures/income-bar.png')
```

# Scatter plot - used for random values

```python
### Problem 4:
##### Problem statement
Generate a Scatter plot for fifty random numbers.k
#### Constraints

#### Test Cases
* Test Case 1
* Test Case 2
```

```python
N = 50 
x = np.random.rand(N) #generate 50 random numbers
y = np.random.rand(N) #generate 50 random numbers
colorlist = np.random.rand(N) #generate 50 random color values
arealist = np.pi * (15 * np.random.rand(50))**2  #generate 50 random areas of circles

plt.scatter(x,y,s = arealist,c = colorlist,alpha = 5)

# alpha values corresponds to the intensity of color.more is the alpha ,more is the intensity and vice-versa.

plt.savefig('python figures/scatterplot-50-randomnumbers.png')
```

```python
import matplotlib.pyplot as plt

# Data to plot
labels = 'distinction', 'first class', 'second class', 'Failed'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightskyblue', 'lightcoral']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```
